```python
import optuna
```

```python
name = "GDSC2_GAT"
study = optuna.create_study(
    storage="sqlite:///{}.sqlite3".format(name),
    study_name=name,
    load_if_exists=True,
)
res = study.trials_dataframe()
res.dropna(subset=[i for i in res.columns if "values" in i])
```

```python
name = "GDSC2"
study = optuna.create_study(
    storage="sqlite:///{}_{}.sqlite3".format(name, "GCN_MPNN"),
    study_name=name,
    load_if_exists=True,
)
res = study.trials_dataframe()
res.dropna(subset=[i for i in res.columns if "values" in i])
```
