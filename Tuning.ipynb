```python
import optuna
import pandas as pd
from metrics import get_parsed_df
```

```python
name = 'NCI'
method = "GATv2"  # 任意のstudy名
storage_path = f"./Tuning_{name}/{method}.sqlite3"  # "./subdir"は1階層下のディレクトリ
study = optuna.create_study(
    storage=f"sqlite:///{storage_path}",
    study_name=method,
    load_if_exists=True,
)

tmp = study.trials_dataframe().dropna(subset=['values_0', 'values_1', 'values_2', 'values_3'])
tmp = tmp[[i for i in tmp.columns if 'params' in i]]
df = (study.trials_dataframe().dropna(subset=['values_0', 'values_1', 'values_2', 'values_3']).iloc[:, 21:-2])
df.columns = [i.replace('user_attrs_', '') for i in df.columns]
print(df.shape)
parsed_df = get_parsed_df(df)
parsed_df = pd.concat([parsed_df, tmp], axis=1)
parsed_df.sort_values('AUPR', ascending=False).head()
```
