```python
import numpy as np
import optuna
import pandas as pd
```

```python
name = "CTRP_GAT"
study = optuna.create_study(
    storage="sqlite:///{}.sqlite3".format(name),
    study_name=name,
    load_if_exists=True,
)
df = study.trials_dataframe().dropna(
    subset=["values_0", "values_1", "values_2", "values_3"]
)
df
```

```python
tmp = df.loc[
    df[["values_0", "values_1", "values_2", "values_3"]]
    .max(axis=1)
    .sort_values(ascending=False)
    .index
]
tmp[tmp["params_gnn_layer"] == "GAT"]
```

```python
tmp[tmp["params_gnn_layer"] == "GATv2"]
```

```python
tmp[tmp["params_gnn_layer"] == "Transformer"]
```

```python
df.params_gnn_layer.value_counts()
```

```python
name = "CTRP"
study = optuna.create_study(
    storage="sqlite:///{}_{}.sqlite3".format(name, "GCN_MPNN"),
    study_name=name,
    load_if_exists=True,
)
df = study.trials_dataframe().dropna(
    subset=["values_0", "values_1", "values_2", "values_3"]
)
df.sort_values(["values_0", "values_1", "values_2", "values_3"], ascending=False)
```
