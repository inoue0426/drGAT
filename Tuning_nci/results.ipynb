```python
import optuna
```

```python
name = "NCI_GAT"
study = optuna.create_study(
    storage="sqlite:///{}.sqlite3".format(name),
    study_name=name,
    load_if_exists=True,
)
df = study.trials_dataframe()
df = df.dropna(subset=[i for i in df.columns if "values" in i])
df
```

```python
df.params_gnn_layer.value_counts()
```
