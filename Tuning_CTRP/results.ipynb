```python
import numpy as np
import optuna
import pandas as pd
```

```python
name = "CTRP_GAT"
study = optuna.create_study(
    storage="sqlite:///{}.sqlite3".format(name),
    study_name=name,
    load_if_exists=True,
)
study.trials_dataframe().dropna(subset=["values_0", "values_1", "values_2", "values_3"])
```

```python
name = "CTRP"
study = optuna.create_study(
    storage="sqlite:///{}_{}.sqlite3".format(name, "GCN_MPNN"),
    study_name=name,
    load_if_exists=True,
)
df = study.trials_dataframe().dropna(
    subset=["values_0", "values_1", "values_2", "values_3"]
)
df = df[
    ~(
        (df["values_0"] == float("-inf"))
        | (df["values_1"] == float("-inf"))
        | (df["values_2"] == float("-inf"))
        | (df["values_3"] == float("-inf"))
    )
]
df.sort_values(["values_0", "values_1", "values_2", "values_3"], ascending=False)
```
